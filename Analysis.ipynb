```python
import pandas as pd

data = pd.read_csv('./interaction_features.csv')
print('Data points: ' + str(len(data)))
data
```

```python
avg_fmeasure = pd.to_numeric(data[data['f.measure'] != 'null']['f.measure']).mean()
avg_fmeasure
```

```python
labels = data[data.columns[5:9]]
labels = labels.fillna(value=0)
labels = labels.replace(to_replace='null', value=-1)
labels.head()
```

```python
features = data[data.columns[12:]]
features = features.fillna(value=0)
features.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.ensemble import ExtraTreesClassifier

forest = ExtraTreesClassifier(n_estimators=250, random_state=4)

forest.fit(features, accurate)
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(features.shape[1]):
    print("%d. feature %d %s (%f)" % (f + 1, indices[f], features.columns[indices[f]], importances[indices[f]]))
```

```python
plt.figure()
plt.title("Feature importances")
plt.bar(range(features.shape[1]), importances[indices],
       color="r", yerr=std[indices], align="center", )
plt.xticks(range(features.shape[1]), indices)
plt.xlim([-1, features.shape[1]])
plt.show()
```

```python
import_df = pd.DataFrame(data={'feature': features.columns, 'importance': importances})
import_df.head()
```

```python
types_of_feature = ['total', 'per_pan_mean', 'per_pan_std']
feature_type_accuracy = pd.DataFrame(columns=['type', 'mean', 'std'])
for t in types_of_feature:
    df = import_df[import_df['feature'].str.endswith(t)]
    feature_type_accuracy.loc[len(feature_type_accuracy)] = [t, df['importance'].mean(), df['importance'].std()]
feature_type_accuracy
```

```python
plt.bar(feature_type_accuracy['type'], feature_type_accuracy['mean'], yerr=feature_type_accuracy['std'], align="center")
plt.title('Importance of types of features')
plt.show()
```

```python
for t in types_of_feature:
    df = import_df[import_df['feature'].str.endswith(t)]
plt.violinplot([import_df[import_df['feature'].str.endswith(t)]['importance'] for t in types_of_feature], showmedians=True)
plt.title('Importance of types of features')
plt.xticks(range(1, 4), types_of_feature)
plt.show()
```
